## Select the names of all employees in Boston 

Hint: join the employees and offices tables. Select the first and last name.

```python
# select the name of all employees in Boston
query = """
SELECT firstName,lastName
FROM employees
JOIN offices
USING (officecode)
WHERE city = "Boston";
"""
pd.read_sql(query,conn)
```

## Are there any offices that have zero employees?
Hint: Combine the employees and offices tables and use a group by. Select the office code, city, and number of employees.

```python
# check if there is any offices with zero employees
query = """
SELECT officeCode,city,COUNT(employeeNumber) AS num_employees
FROM offices
JOIN employees
USING (officecode)
GROUP BY officeCode
HAVING num_employees=0;
"""
pd.read_sql(query,conn)
```

## Level Up 1: Display the names of every individual product that each employee has sold

Hint: You will need to use multiple `JOIN` clauses to connect all the way from employee names to product names.

```python
# Displaying the names of every individual product that each employee has sold
query ="""SELECT firstName,lastName,productName
         FROM employees AS emp
         JOIN customers AS cust
         ON emp.employeeNumber = cust.salesRepEmployeeNumber
         JOIN orders AS ord
         USING (customerNumber)
         JOIN orderdetails
         USING (orderNumber)
         JOIN products
         USING (productCode);"""
pd.read_sql(query,conn)
```

## Level Up 2: Display the number of products each employee has sold

Alphabetize the results by employee last name.

Hint: Use the `quantityOrdered` column from `orderDetails`. Also, think about how to group the data when some employees might have the same first or last name.

```python
# Display the number of products each employee has sold
query = """SELECT firstName,lastName,SUM(quantityOrdered)AS total_ordered
           FROM employees AS emp
         JOIN customers AS cust
         ON emp.employeeNumber = cust.salesRepEmployeeNumber
         JOIN orders AS ord
         USING (customerNumber)
         JOIN orderdetails
         USING (orderNumber)
         GROUP BY firstName,lastName
         ORDER BY lastName
         """
pd.read_sql(query,conn)
```

## Level Up 3: Display the names employees who have sold more than 200 different products

Hint: this is different from the previous question because the quantity sold doesn't matter, only the number of different products

```python
# Display the names employees who have sold more than 200 different products
query= """SELECT firstName,lastName,COUNT(productCode)AS num_of_products
           FROM employees AS emp
         JOIN customers AS cust
         ON emp.employeeNumber = cust.salesRepEmployeeNumber
         JOIN orders AS ord
         USING (customerNumber)
         JOIN orderdetails
         USING (orderNumber)
         GROUP BY firstName,lastName
         HAVING num_of_products>200
         ORDER BY lastName;"""
pd.read_sql(query,conn)
```
